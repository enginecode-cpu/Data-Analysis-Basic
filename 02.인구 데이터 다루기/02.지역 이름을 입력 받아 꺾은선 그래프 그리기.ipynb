### 지역 이름을 입력 받아 꺾은 선 그래프 그리기

```python
import matplotlib.pyplot as plt
import csv
```

```python
user_location = input()

f = open('./age.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = []
populations = []
location = ''
for v in data:
    if user_location in v[0]:
        location = v[0]
        populations = v[3:]

x = [i for i in range(0, 101)]
y = [int(i.replace(',', '')) for i in populations]

plt.rcParams['font.family'] = 'NanumGothic'
plt.plot(x, y)
plt.title(f'{location}의 인구')
plt.show()
```

```python
plt.bar(x, y)
plt.rcParams['font.family'] = 'NanumGothic'
plt.title(f'{location}의 인구')
plt.show()
```
