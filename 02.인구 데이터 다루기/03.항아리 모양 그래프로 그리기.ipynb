### 항아리 모양 그래프로 그리기 

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('./gender.csv', encoding='cp949')

for i, col in enumerate(df.columns):
    print(i, col)
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

male_x = []
female_x = []
for i, row in enumerate(data):
    if i == 1789:
        male_x = row[3:104]
        female_x = row[106:]
```

```python
mx = [-int(m_age.replace(',','')) for m_age in male_x]
fmx = [int(fm_age.replace(',','')) for fm_age in female_x]
y = [i for i in range(0, 101)]
```

```python
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['axes.unicode_minus'] = False
plt.title('경기도 광주시 지역의 남녀 성별 인구 분포')
plt.barh(y, mx, label='남성')
plt.barh(y, fmx, label='여성')
plt.legend()
plt.show()
```
