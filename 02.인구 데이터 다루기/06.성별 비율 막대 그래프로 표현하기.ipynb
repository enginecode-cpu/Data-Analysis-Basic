### 성별 비율 막대 그래프로 표현하기

```python
import matplotlib.pyplot as plt
import csv
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

user_location = input()

male_y = []
female_y = []
location = ''
for row in data:
    if user_location in row[0]:
        location = row[0]
        male_y = row[3:104]
        female_y = row[106:]

x = [i for i in range(0, 101)]
y = []
for m, fm in zip(male_y, female_y):
    y.append(int(m.replace(',', '')) - int(fm.replace(',', '')))

plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.title(f'{location} 지역의 남녀 성별 인구 분포')
plt.bar(x, y)
plt.show()
```
