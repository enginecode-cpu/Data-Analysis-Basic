### 제주도 성별 인구 산점도 그리기

```python
import matplotlib.pyplot as plt
import math
import csv
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

jeju_index = 0
jeju_male_x = []
jeju_female_x = []
for i, row in enumerate(data):
    if '제주' in row[0]:
        jeju_index = i
        jeju_male_x = row[3:104]
        jeju_female_x = row[106:]
        break

size = []
for m, f in zip(jeju_male_x, jeju_female_x):
    size.append(math.sqrt(int(m) + int(f)))
```

```python
m_max = max([int(m_age) for m_age in jeju_male_x])
m = [int(m_age) for m_age in jeju_male_x]
fm = [int(f_age) for f_age in jeju_female_x]
x = [i for i in range(101)]
```

```python
plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.title('제주도 지역의 남녀 성별 인구 산점도')
plt.scatter(m, fm, s=size, c=range(0, 101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```

```python
import numpy as np

f = open('gender.csv', encoding='cp949')
data =csv.reader(f)
next(data)

m = []
f = []

for v in data:
    if '제주특별자치도  (5000000000)' in v[0]:
        m = v[3:104]
        f = v[106:]

x = np.arange(0,101)
m_r = [int(a.replace(',','')) for a in m]
f_r = [int(b.replace(',','')) for b in f]

m1 = np.array(m_r)
f1 = np.array(f_r)

result = m1 + f1
# result

size = []
for d in result:
    size.append(math.sqrt(d))

plt.figure(figsize=(10, 5))
plt.scatter(m_r, f_r, s=size, c=x ,alpha=0.5, cmap='jet') # c : 컬러
plt.colorbar()
plt.title('제주도 지역의 남녀 성별 인구 산점도')
plt.plot(range(max(m_r)), range(max(m_r)), 'g')
plt.show()
```
