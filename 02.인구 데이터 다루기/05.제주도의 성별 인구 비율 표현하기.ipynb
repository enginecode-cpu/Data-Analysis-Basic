### 제주도의 성별 인구 비율을 파이차트로 그리기

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('./gender.csv', encoding='cp949')
df
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

jeju_index = 0
jeju_all_male = 0
jeju_all_female = 0
for i, row in enumerate(data):
    if '제주' in row[0]:
        jeju_index = i
        jeju_all_male = int(row[1])
        jeju_all_female = int(row[104])
        break
```

```python
ratio = [jeju_all_male, jeju_all_female]
labels = ['남자', '여자']
```

```python
plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 14
plt.title('제주도 지역의 남녀 성별 인구 분포')
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
