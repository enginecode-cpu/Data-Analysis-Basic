### 우리 동네 인구 구조 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('./age.csv', encoding='cp949')
df
```

```python
f = open('./age.csv', encoding='cp949')
data = csv.reader(f)
next(data)

result = []

for i, row in enumerate(data):
    if i == 1790:
        result = row[3:]

y = [int(r.replace(',', '')) for r in result]
```

```python
x = [i for i in range(0, 101)]
```

```python
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.title('경기도 광주시 나이별 인구')
plt.plot(x, y)
plt.show()
```
