### 제주도에는 여성의 비율이 높을까?

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
f = open('./gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

jeju_index = 0
jeju_male_x = []
jeju_female_x = []
for i, row in enumerate(data):
    if '제주' in row[0]:
        jeju_index = i
        jeju_male_x = row[3:104]
        jeju_female_x = row[106:]
        break
```

```python
mx = [-int(m_age) for m_age in jeju_male_x]
fmx = [int(f_age) for f_age in jeju_female_x]
y = [i for i in range(101)]
```

```python
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.title('제주도 지역의 남녀 성별 인구 분포')
plt.barh(y, mx, label='남성')
plt.barh(y, fmx, label='여성')
plt.legend()
plt.show()
```

```python
my = [int(m_age) for m_age in jeju_male_x]
fmy = [int(f_age) for f_age in jeju_female_x]
x = [i for i in range(101)]
```

```python
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.title('제주도 지역의 남녀 성별 인구 분포')
plt.plot(x, my, label='남성')
plt.plot(x, fmy, label='여성')
plt.legend()
plt.show()
```

```python
my.sort()
fmy.sort()
```

```python
plt.figure(figsize=(10, 5))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.subplot(121)
plt.title('제주도 남자 인구 비율')
plt.boxplot(my)

plt.subplot(122)
plt.title('제주도 여자 인구 비율')
plt.boxplot(fmy)
plt.show()
```
