```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./seoul.csv', encoding='cp949')
df
```

```python
df.info()
```

```python
df = df[df[df.columns[2]].isna() == False]
df
```

```python
new_df = pd.concat([df], ignore_index=True)
new_df.drop(axis=1, columns=[new_df.columns[1], new_df.columns[3], new_df.columns[4]], inplace=True)
new_df
```

```python
date = new_df[new_df.columns[0]]
temp = new_df[new_df.columns[1]]

my_birthday = []
my_birthday_temp = []
for i, (d, t) in enumerate(zip(date, temp)):
    if d.split('-')[1] == '10' and d.split('-')[2] == '02':
        my_birthday.append(d)
        my_birthday_temp.append(t)
```

```python
plt.figure(figsize=(10,5))
plt.plot(my_birthday, my_birthday_temp)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(my_birthday, my_birthday_temp)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```
