```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

date = []
result = []

for row in data :
    if row[2] != '':
        if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '02':
            result.append(float(row[2]))
            date.append(row[0])
```

```python
plt.plot(date, result)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []

for row in data :
    if row[-1] != '':
        result.append(float(row[-1]))
```

```python
result.sort()
```

```python
plt.boxplot(result)
plt.show()
```

```python
result_nparray = np.array(result)

print(f'min: {np.percentile(result_nparray, 0)}')
print(f'1/4: {np.percentile(result_nparray, 25)}')
print(f'2/4: {np.percentile(result_nparray, 50)}')
print(f'3/4: {np.percentile(result_nparray, 75)}')
print(f'max: {np.percentile(result_nparray, 100)}')
```
