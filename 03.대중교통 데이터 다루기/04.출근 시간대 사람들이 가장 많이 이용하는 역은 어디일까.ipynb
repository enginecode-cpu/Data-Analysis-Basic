### 출근 시간대 사람들이 가장 많이 이용하는 역은 어디일까?

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./subwaytime.csv', encoding='cp949')
df
```

```python
f = open('./subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

result = []
for i, row in enumerate(data):
    row[4:] = map(int, row[4:])
    result.append((row[3], row[10]))
```

```python
result.sort(key=lambda x: x[1])

x = []
y = []
for (xvalue, yvalue) in result:
    x.append(xvalue)
    y.append(yvalue)
```

```python
plt.figure(figsize=(100,50))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.bar(x, y)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```

```python
f = open('./subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

result = []
for i, row in enumerate(data):
    row[4:] = map(int, row[4:])
    result.append((row[3], sum(row[10:15:2])))
```

```python
x = []
y = []
for (xvalue, yvalue) in result:
    x.append(xvalue)
    y.append(yvalue)
```

```python
plt.figure(figsize=(100,50))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.bar(x, y)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```

```python
sorted_result = sorted(result, key=lambda x: x[1], reverse=True)
mx_station, mx = sorted_result[0][0], sorted_result[0][1]
```

```python
print(f'출근 시간에 {mx_station}역에서 {mx}로 가장 많이 이용합니다.')
```
