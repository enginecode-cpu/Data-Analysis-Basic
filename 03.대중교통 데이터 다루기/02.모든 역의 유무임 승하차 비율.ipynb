### 모든 역의 유무임 승하차 비율

```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('./subwayfee.csv', encoding='cp949')
data = csv.reader(f)
next(data)

label = ['유임승차', '유임하차', '무임승차', '무임하차']

non_free_riding = []
non_free_quit = []
free_riding = []
free_quit = []

for row in data:
    for i in range(4,8):
        if i == 4:
            non_free_riding.append(int(row[i]))
        elif i == 5:
            non_free_quit.append(int(row[i]))
        elif i == 6:
            free_riding.append(int(row[i]))
        elif i == 7:
            free_quit.append(int(row[i]))
```

```python
ratio = [sum(non_free_riding), 
         sum(non_free_quit), 
         sum(free_riding), 
         sum(free_quit)]
```

```python
plt.figure(figsize=(10,5))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.pie(ratio, labels=label, autopct='%.1f%%')
plt.show()
```
