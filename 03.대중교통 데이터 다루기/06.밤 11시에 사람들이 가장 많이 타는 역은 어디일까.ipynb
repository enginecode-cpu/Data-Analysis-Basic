### 밤 11시에 사람들이 가장 많이 이용하는 역은 어디일까?

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df = pd.read_csv('./subwaytime.csv', encoding='cp949')
df.columns
```

```python
f = open('./subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

t = int(input())

result = []
i = 4 + (t - 4) * 2
for row in data:
    row[4:] = map(int, row[4:])
    result.append((row[3], row[i]))

result.sort(key=lambda x: x[1])

x = []
y = []
for (xvalue, yvalue) in result:
    x.append(xvalue)
    y.append(yvalue)

plt.figure(figsize=(100,50))
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.bar(x, y)
plt.xticks(rotation=90)
plt.tick_params(axis='x', labelsize=4)
plt.show()
```

```python
sorted_result = sorted(result, key=lambda x: x[1], reverse=True)
mx_station, mx = sorted_result[0][0], sorted_result[0][1]
```

```python
print(f'{t}시에 {mx_station}역에서 {mx}로 가장 많이 탑니다.')
```
